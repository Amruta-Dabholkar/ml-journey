Matplotlib Basics

```python
x = np.linspace(0,5,11)
x
y = x**3
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("Chacha")
plt.ylabel("Chachi")
```

Subplots in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```
