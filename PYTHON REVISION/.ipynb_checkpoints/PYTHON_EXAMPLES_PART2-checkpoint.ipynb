IF - ELSE CONDITIONS AND DECISION MAKING

Basic if-else

```python
age = 18
if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")
```

if-elif-else chain

```python
marks = 75
if marks >= 90:
    grade ="A+"
elif marks >= 80:
    grade ="A"
elif marks >= 70:
    grade ="B"
elif marks >= 60:
    grade = "C"
else:
    grade="Fail"
print(f"Grade:{grade}")
```

Nested if

```python
age = 20
income = 50000

if age >= 18:
    if income >= 30000:
        print("Eligibile for loan")
    else:
        print("Age OK but income too low")
else:
    print("Too young for loan")
```

Ternary (one-line-if-else)

```python
x=10
result = "Even" if x%2 == 0  else "Odd"
```

```python
age = 20
status = "Adult" if age >= 18 else "Minor"
```

Practical Example

```python
#BMI calculator 
weight = float(input("Weight(kg):"))
height = float(input("Height(m):"))
bmi = weight/(height**2)
```

```python
if bmi < 18.5:
    category  = "Underweight"
elif bmi < 25:
    category = "Normal"
elif bmi < 30:
    category = "Overweight"
else:
    category = "Obese"
```

```python
print(f"BMI:{bmi:.1f}-{category}")
#Login system
username = "admin"
password="1234"
u = input("Username:")
p = input("Password:")
```

```python
if u == username and p == password:
    print("Login successful!")
elif u == username:
    print("Wrong password")
else:
    print("User not found")
```

```python
#Calculator
a = float(input("Enter first number:"))
op = input("Enter operator(+,-,*,/):")
b = float(input("Enter second number:"))
```

```python
if op == "+":
    print(f"Result:{a+b}")
elif op == "-":
    print(f"Result:{a_b}")
elif op == "*":
    print(f"Result:{a*b}")
elif op == "/":
    print(f"Result:{a/b}")
else:
    print("Invalid Operator")
```
