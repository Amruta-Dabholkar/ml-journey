Default arguments

```python
def greet(name,msg="Good Morning"):
    print(f"{msg},{name}!")
greet("Rohan")
greet("Rohan","Good Night!")
```

Keyword arguments

```python
def student_info(name,age,marks):
    print(f"Name:{name},Age:{age},Marks:{marks}")
student_info("Rohan",21,90)
student_info(marks=90,name="Rohan",age=21)
```

args - variable number of arguments

```python
def add_all(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total
print(add_all(1,2,3))
print(add_all(1,2,3,4,5))
print(add_all(10,20))
```

kwargs — variable keyword arguments

```python
def display(**info):
    for key,value in info.items():
        print(f"{key}:{value}")
display(name="Rohan",age=21,city="Mumbai")
```

Return multiple values

```python
def min_max(numbers):
    return min(numbers),max(numbers)
low,high=min_max([3,1,7,2,9])
print(low,high)
```

Scope - local vs global

```python
x="global"
def show():
    x="local" # local variable, doesn't affect global 
    print(x) #local

show()
print(x) #global

#global keyword
count = 0
def increment():
    global count
    count += 1
increment()
increment()
print(count)
    
```

Lambda functions (one-line anonymous functions)

```python
square = lambda x:x**2
print(square(5))
```

```python
add = lambda a,b:a+b
print(add(3,4))
```

```python
# Lambda with map, filter, sorted
```

```python
num = [1,2,3,4,5]

square = list(map(lambda x:x**2,num))
print(square)
```

```python
evens = list(filter(lambda x:x%2 == 0,num))
print(evens)
```

```python
words = ["banana","apple","cherry","date"]
sorted_words = sorted(words,key=lambda x:len(x))
print(sorted_words)
```

Recursive functions

```python
# Factorial
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5))
print(factorial(0))
```

```python
# Fibonacci
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
for i in range(8):
    print(fib(i),end=" ")
```

```python
# Binary search (recursive)
def binary_search(arr,target,low,high):
    if low>high:
        return -1
    mid = (low+high)//2
    if arr[mid]==target:
        return mid
    elif arr[mid]<target:
        return binary_search(arr,target,mid+1,high)
    else:
        return binary_search(arr,target,low,mid-1)
arr = [1,3,5,7,9,11,13]
print(binary_search(arr,7,0,len(arr)-1))
```

Higher-order functions

```python
def apply(func,value):
    return func(value)
def double(x):
    return x*2
print(apply(double,5))
print(apply(lambda x:x**2,4))
```

```python
# Function returning function (closure)
def multiplier(factor):
    def multiply(x):
        return x*factor
    return multiply
double = multiplier(2)
triple = multiplier(3)

print(double(5))
print(triple(5))
```

SET

```python
# Create — unordered, no duplicates
s = {1,2,3,4,5}
s2 = set([1,2,2,3,3,3])   # duplicates removed
print(s2)
```

```python
#methods
s = {1,2,3,4,5}
print(s.add(6))
print(s.remove(1))
print(s.discard(99))
print(len(s))
print(3 in s)
```

```python
# Set operations 
a = {1, 2, 3, 4, 5} 
b = {4, 5, 6, 7, 8}
print(a | b)    # union → {1,2,3,4,5,6,7,8} 
print(a & b)    # intersection → {4, 5} 
print(a - b)    # difference → {1, 2, 3} 
print(a ^ b)    # symmetric difference → {1,2,3,6,7,8} 
```

```python
# Subset / superset
x = {1, 2, 3} 
y = {1, 2, 3, 4, 5}
print(x.issubset(y))
print(y.issuperset(x))
```

```python
# Remove duplicates from list using set 
names = ["Rohan", "Amit", "Rohan", "Priya", "Amit"]
unique = list(set(names))
print(unique)
```

DICTIONARY 

```python
# Create — key: value pairs, keys must be unique
student = {
    "name" : "Rohan",
    "age":21,
    "marks":90.5,
    "subjects":["Math","Python","ML"]
}
```

```python
#access
print(student["name"])
print(student.get("age"))
print(student.get("email","N/A"))
```

```python
# Modify
student['age']=22
print(student.get("age"))
student["email"]="rohan@gmail.com"
print(student.get("email"))
```

```python
#delete
```

```python
student = {"name": "Alex", "email": "alex@example.com"}
del student["email"]
print(student)   # {"name": "Alex"}
```

```python
student = {"name": "Alex", "marks": 85}
popped = student.pop("marks")
print(popped)    # 85
print(student)   # {"name": "Alex"}
```

```python
# Methods 
student = {
    "name" : "Rohan","age":21,"marks":90.5,"subjects":["Math","Python","ML"]
}
print(student.keys())     # dict_keys(['name', 'age', 'subjects']) 
print(student.values())   # dict_values([...]) 
print(student.items())    # dict_items([...])
```

```python
#loop
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
# Check key
print("name" in student)
print("email" in student)
```

```python
# Nested dictionary
college = {
    "CS":{"students":200,"fees":80000},
    "IT":{"students":150,"fees":70000}
}
print(college["CS"]["fees"])
print(college["IT"]["fees"])
```

```python
# Dictionary comprehension
squares = {x:x**2 for x in range (1,6)}
print(squares)
```

```python
even_squares = {x:x**2 for x in range(1,11) if x%2==0}
print(even_squares)
```

```python
# Word count
sentence = "the cat sat on the mat"
word_count = {}
for word in sentence.split():
    word_count[word] = word_count.get(word,0)+1
    print(word_count)
```
