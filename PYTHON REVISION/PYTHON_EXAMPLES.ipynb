```python
#String,Operators & input/output

#string basics
s1="hello"
s2='world'
s3="""This is
a multiline comment"""
print(s1+ " " +s2)
print(s1 * 3)
print(len(s1))
print(s1[0])
print(s1[-1])
print(s1[1:4])
print(s1[::-1]) #reverse
```

```python
#String Methods
s = "Hello World"
print(s.upper())
print(s.lower())
print(s.strip()) #removes space
print(s.replace("World","Python"))
print(s.split())
print(s.find("World"))
print(s.count("l"))
print(s.startswith("He"))
print(s.endswith("ld"))
print("hello".capitalize())
print("hello world".title())
print(",".join(["a","b","c"]))
```

```python
#f-string
name = "Rohan"
age = 21
marks = 95.567
print(f"Name:{name}")
print(f"Age:{age}")
print(f"Marks:{marks:.2f}")
print(f"Is Adult:{age>=18}")
print(f"Name uppercase : {name.upper()}")
```

```python
#OPERATORS
#arithmetic
print(10+3)
print(10-3)
print(10*3)
print(10/3)
print(10//3)
print(10%3)
print(10**3)
```

```python
#comparison
print(5>3)
print(5<3)
print(5==3)
print(5!=3)
print(5>=3)
print(5<=3)
```

```python
#logical
print(True and False)
print(True or False)
print(not True)
print(5>3 and 10>7)
print(5>3 or 10>7)
```
