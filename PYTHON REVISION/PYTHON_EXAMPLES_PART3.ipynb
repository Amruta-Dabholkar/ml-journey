 LOOPS — FOR, WHILE + PATTERN QUESTIONS
 
For loop basics

```python
# Loop over range
for i in range(5):
    print(i)
```

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1,11,2):
    print(i)
```

```python
#reverse
for i in range (10,0,-1):
    print(i)
```

Loop over sequences

```python
# Loop over string
for char in "Python":
    print(char)
```

```python
# Loop over list
fruits = ["apple","banana","mango"]
for fruit in fruits:
    print(fruit)
```

```python
# Loop with index
for i,fruit in enumerate(fruits):
    print(f"{i}:{fruit}")
```

```python
# Loop over dictionary
student = {"name":"Rohan","age":21,"marks":90}
for key, value in student.items():
    print(f"{key}:{value}")
```

While loop

```python
# Basic while
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# While with input
while True:
    answer = input("Continue?(yes/no):")
    if answer == "no":
        break
    print("Continuing...")
```

```python
# Guess the number
import random
secret = random.randint(1,10)
attempts = 0

while True:
    guess = int(input("Guess(1-10):"))
    attempts += 1
    if guess == secret:
        print(f"Correct!Took {attempts} attempts")
        break
    elif guess < secret:
        print("Too low")
    else:
        print("Too high")
```

break, continue, pass

```python
# break → exit loop completely
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue → skip current iteration
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
# pass → do nothing (placeholder)
for i in range(5):
    if i == 3:
        pass
    print(i)
```

Nested loops

```python
# Multiplication table
for i in range(1,6):
    for j in range(1,6):
        print(f"{i * j : 4}",end ="")
    print()
```

Pattern Questions

```python
# Pattern 1: Right triangle
n = 5
for i in range(1,n+1):
    print("*"*i)
```

```python
# Pattern 2: Inverted triangle
for i in range(n,0,-1):
    print("*"*i)
```

```python
# Pattern 3: Number triangle
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
# Pattern 4: Pyramid
for i in range(1,n+1):
    print(" "*(n-i)+"*"*(2*i-1))
```

```python
# Pattern 5: Diamond
for i in range(1, n+1):
 print(" " * (n-i) + "*" * (2*i-1))
for i in range(n-1, 0, -1):
 print(" " * (n-i) + "*" * (2*i-1))
```

```python
# Pattern 6: Floyd's triangle
num = 1
for i in range(1,n+1):
    for j in range (i):
        print(num,end=" ")
        num += 1
    print()
```

```python
# Pattern 7: Pascal's triangle
from math import comb
for i in range(n):
 print(" " * (n-i), end="")
 for j in range(i+1):
     print(comb(i,j), end=" ")
 print()
```

Loop examples 

```python
# Sum of digits
n = 1234
total = 0
while n > 0:
    total += n%10
    n //= 10
    print(f"Sum of digit:{total}")
```

```python
# Check prime number
def is_prime(n):
    if n<2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return False
    return True
print(is_prime(17)) #true
print(is_prime(10)) #false
# Print all primes up to 50
primes = [ n for n in range(2,51) if is_prime(n)]
print(primes)
```

```python
# Fibonacci series
a = 0
b = 1
for _ in range(10):
    print(a,end=" ")
    a,b=b,a+b
```

```python
# Factorial
n = 6
fact = 1
for i in range(1,n+1):
    fact *= i
print(f"{n}!={fact}")
```

```python
# Reverse a number
n = 12345
rev = 0
while n>0:
    rev = rev * 10 + n % 10
    n//=10
print(rev)
```
