Numpy Indexing And Slicing Of Matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[1]
```

```python
arr[5]
```

```python
arr[0,0]
```

```python
arr[0,4]
```

```python
arr[5,4]
```

```python
slice = arr[0:2,1:3]
slice
```

```python
slice_2 = arr[3: ,3: ]
slice_2
```

```python
slice_3 = arr[:,2]
slice_3
```

Boolean Indexing

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
